# Estrutura de Repetição


```python
# example with while
count = 1
while count <= 5:
  print(count)
  count +=1
```

```python
# range (0, 10, 1):
# range (start, stop, increment)
# the python "ignores" the last value of intervals
#The increment may be negative

# for normal
for n in range(0, 10):
  print(n+1)

for n in range(1, 11):
  print(n)
```

```python
#for decrescente

for n in range(10, 0, -1):
  print(n)
```

```python
#example with break
# will stop at 3 because the break is before the print
for n in range(0, 10):
  if n ==4:
    break
  print(n)
```

```python
#example with continue
# will skip number 4
for n in range(0, 10):
  if n == 4:
    continue
  print(n)
```

# Numpy

```python
import numpy as np
```

```python
#create a single-dimension array;

mt = np.array([12, 34, 56, 75, 10, 18])
print(mt)
print(type(mt))
```

```python
# create the array with a specific type
# create the array as 64-bit float

mtfloat = np.array([1, 2, 3], dtype = np.float64)
print(mtfloat)
print(type(mtfloat))
mtint =  np.array([1, 2, 3], dtype = np.int32)
print(mtint)
print(type(mtint))
```

```python
#change the type of the array
#we can transform data types of arrays

mtnew = np.array([1.4, 3.6, -5.1, 9.42, 4.999999])
print(mtnew)
# when we transform float to int the values are truncated
mtnewint = mtnew.astype(np.int32)
print(mtnewint)
```

```python
#we can do the reverse as well
mt5 = np.array([1, 2, 3, 4, 5])
print(mt5)
mt6 = mt5.astype(float)
print(mt6)
```

```python
# more than one dimension
# creates a two-dimensional array

mt7 = np.array([[1, 2, 4], [12, 27, 3], [8, 9, 53]])
print(mt7)
```

```python
#create typed empty arrays
# empty means that they are not initialized, not that they are empty
vazio = np.empty([3, 2], dtype = int)
print(vazio)
print("---------")

#creates a 4x3 matrix with zero values
zeros = np.zeros([4, 3])
print(zeros)
print('---------')

#with values equal to 1
one = np.ones([5, 7])
print(one)
print('--------')

#creates square matrix with main diagonal with values 1 and other values zero.
diagonal = np.eye(5)
print(diagonal)
```

```python
# random values between 0 and 1
ale = np.random.random((5))
print(ale)
print('-----------------------------------------------')

# random values normal distribution containing negative
ale2 = np.random.randn((5))
print(ale2)
print('-----------------------------------------------')

# random values 3 x 4
ale3 = (10*np.random.random((3, 4)))
print(ale3)
```

```python
#another way to generate random
#use of seed
gnr = np.random.default_rng(1)
ale5 = gnr.random(3)
print(ale5)
print('-----------------------------------------------')

#generate integers
ale6 = gnr.integers(10, size=(5, 4))
print(ale6)
```

```python
# unique removes repetitions
j= np.array([12, 12, 13, 14, 15, 16, 17, 12, 13, 11, 18, 19, 20])
j = np.unique(j)
print(j)
```

```python
# specific functions
# create the two-dimensional matrix k
k = np.array([[17, 22, 43], [27, 25, 14], [15, 24, 32]])

# display the matrix k
print(k)

# show a specific element of the matrix k
print(k[0][1])

# show the size of the dimensions of the matrix k
print(k.shape)
```

```python
print(k)
# Mathematical Functions
# Show the largest value of the matrix k
print(k.max())
# Show the smallest value of matrix k
print(k.min())
# Shows the sum of the values of the matrix k
print((k.sum()))
# Show the average value of the values of the matrix k
print(k.mean())
# Show the value of the standard deviation of the values of the matrix k
print(k.std())
```

```python
# Universal functions, applied to all elements
k1 = np.array([1, 4, 9, 16, 25, 36])
# Show the square root value of all elements
print(np.sqrt(k1))
# Show the exponential value of all elements
print(np.exp(k1))
```

```python
# Extraction of elements
m = np.array ([1, 2, 3, 4, 5, 6])
# Show the element of position 2
print(m[1])
# Show the array created from position 0, two elements
print(m[0:2])
print('------------')
# Show the array created from position 2 to the rest of the array
print(m[1:])
print('------------')
# Show the array created from the ante penultimate position to the end.
print(m[-3:])
```

```python
# Extraction of rows and columns
l = np.array([[4, 5], [6, 1], [7, 4]])
print(l)
print('---------')

# First row, all columns
l_row_1 = l[0, :]
print(l_row_1)
print('---------')

# Second row
l_row_2 = l[1, :]
print(l_row_2)
print('---------')

#Third row
l_row_3 = l[2, :]
print(l_row_3)
print('---------')

# All rows, first column
l_column_1 = l[:, 0]
print(l_column_1)
print('---------')

# All rows, second column
l_column_2 = l[:, 1]
print(l_column_2)
```

```python
#Addition and multiplication of arrays
n = np.array([[1, 2], [3, 4]])
o = np.array([[1, 1], [1, 1]])

resultado1 = n + o
print(resultado1)
print('----------')

resultado2 = n * o
print(resultado2)
print('----------')

p = np.array([[1, 2], [3, 4], [5, 6]])
q = np.array([[2, 1]])
resultado3 = p + q
print(resultado3)
```

```python
# Transposition, rearranges a set of 15 elements from 0 to 14 into 3 rows and 5 columns
f = np.arange(15).reshape((3, 5))
#Show the transposed array between row and column
print(f)
print('-----------')
s = f.T
print(s)
```

```python
# Another way to do it, same result
r = np.arange(15).reshape((3,5))
print(r)
# Rearrange a set of3 15 elements
# Show the matrix transposed between row and column
s = r.transpose((1, 0))
print(s)
```

```python
#Logical expressions
# Using where
# Creating array with positive and negative random values.
v = np.random.randn(4, 4)
print(v)
print('------------')

# Creating array with boolean values based on array v
x = (v > 0)
print(x)
print('----------')

# Creating array with values -1 and 1 based on array x values
z = np.where(x > 0, 1, -1)
print(z)
```
